***5.What are iterators in Python and how do they differ from iterables?***

Ans. Iterators and iterables are two fundamental concepts in Python:

Iterable:

An iterable is an object that can return its elements one at a time, allowing you to iterate over them. Examples include:

- Lists
- Tuples
- Sets
- Dictionaries
- Strings

Iterables have the __iter__ method, which returns an iterator object.

Iterator:

An iterator is an object that keeps track of its position in an iterable and returns the next element on each iteration. Iterators have the __next__ method, which returns the next element.

Key differences:

1. Purpose: Iterables provide access to their elements, while iterators navigate through those elements.
2. State: Iterables maintain their state, whereas iterators keep track of their position.
3. Creation: Iterables create iterators using the __iter__ method.



```python
#Examples
# Iterable (list)
my_list = [1, 2, 3, 4, 5]

# Create an iterator from the iterable
my_iterator = iter(my_list)

# Iterate using the iterator
print(next(my_iterator))

```

```python
print(next(my_iterator))
```

```python
print(next(my_iterator))
```

```python
# Iterate directly over the iterable
for element in my_list:
    print(element)
```

***6.  Explain the concept of generators in Python and how they are defined.***

Ans. Generators are a type of iterable, like lists or tuples, but they don't store all values in memory at once.
They generate values on-the-fly as you iterate over them.
*Useful for large datasets, infinite sequences, or computationally expensive calculations.

Defining Generators:

_Use the yield keyword_ in a function to define a generator.
*yield pauses the function, returning a value, and resumes where it left off.

Examples:-

```python
def infinite_sequence():
    num = 0
    while True:
        yield num
        num += 1

gen = infinite_sequence()
for _ in range(10):
    print(next(gen))
```

***7. What are the advantages of using generators over regular functions?***

Ans.1. Memory Efficiency

Generators store only one value at a time, reducing memory usage.

Example:



```python
# Regular function
def get_squares(n):
    return [x**2 for x in range(n)]  # Stores entire list in memory


get_squares(5)
```

```python
# Generator function
def get_squares_gen(n):
    for x in range(n):
        yield x**2  # Stores only one value at a time

get_squares_gen(5)
```

```python
gen = get_squares_gen(5)
gen
```

2. Lazy Evaluation

Generators compute values on-the-fly, reducing overhead.

Example:





```python
# Regular function
def fibonacci(n):
    fib_sequence = [0, 1]
    for i in range(2, n):
        fib_sequence.append(fib_sequence[i-1] + fib_sequence[i-2])
    return fib_sequence

fibonacci(10)
```

```python
# Generator function
def fibonacci_gen(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

fibonacci_gen(10)
```

```python
fibon = fibonacci_gen(10)
fibon
```

3. Flexibility

Generators can handle infinite sequences.

Example:



```python
# Generator function
def infinite_sequence():
    num = 0
    while True:
        yield num
        num += 1


gen = infinite_sequence()
for _ in range(10):
    print(next(gen))
```

4. Improved Performance

Generators reduce function call overhead.

Example:



```python
# Regular function
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

factorial(5)
```

```python
# Generator function
def factorial_gen(n):
    result = 1
    for i in range(1, n+1):
        yield result
        result *= i

for fact in factorial_gen(10):
    print(fact)
```

***9. Explain the purpose and usage of the `map()` function in Python.***

Ans. The map() function in Python applies a given function to each item of an iterable (such as a list, tuple, or string) and returns a list of the results.

Purpose:

1. Data Transformation: Apply a function to transform each element of an iterable.
2. Data Processing: Perform operations on each element, like filtering, converting, or calculating.
3. Code Simplification: Replace loops with a concise, readable, and efficient way to process iterables.


Usage:

map(function, iterable)

- function: The function to apply to each element.
- iterable: The list, tuple, string, or other iterable to process.

Let's dive into some examples of map() function

```python
#Square numbers:
numbers = [1, 2, 3, 4, 5]
list(map(lambda x: x**2, numbers))

```

```python
#Convert strings to uppercase:
fruits = ["apple", "banana", "cherry"]
list(map(lambda x: x.upper(), fruits))
```

```python
#capitalizing first letter
names = ["john", "alice", "bob"]
list(map(lambda x: x.capitalize(), names))
```

***4. Write a Python function that checks if a given number is prime or not from 1 to 200.***

```python
def check_prime(n):
    if n <= 1:
        return f"{n} is not a prime number"
    for i in range(2,n):
        if n % i == 0:
            return f"{n} is not a prime number"
    return f"{n} is a prime number"

num = int(input("Enter a number between 1 and 200: "))
print(check_prime(num))
```

***6. Write a generator function in Python that yields the powers of 2 up to agivenexponent.***

***10. Create a Python program that uses `filter()` to remove all the vowels from a given string.***

```python
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda c: c not in vowels, s))

input_string = "Hello World!"
print("Original String:", input_string)
print("String without vowels:", remove_vowels(input_string))


```
